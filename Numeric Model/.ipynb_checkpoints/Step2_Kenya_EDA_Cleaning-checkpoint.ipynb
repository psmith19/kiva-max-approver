## The plan is to pick Kenya (CountryCode = KE) given the higher volumes at 51k and relatively lower funded rate of 78%.  Basically the baseline expectation is 78% for funded rate.

```python
#Creating a subset of accounts for Kenya
kivake = kiva[kiva['COUNTRY_CODE']=='KE']
kivake.shape
```

```python
# See missing variables
kivake.isna().sum()
```

### Key Observations:  
#1 Missing variables for text data in Description, Tags and Loan Use.  
#2 Also not everyone has provided images and videos.  The IDs themselves are not as valuable here, however, the presence of images and videos is valuable.
#3 Multiple times variables.  Raised time is missing as expected since non-funded will not have that.  Posted-time is present in all as expected and will be used for feature engineering the month to see if there is seasonality there.

```python
#Converting long text columns to length of words and count of characters
#Step1 is to replace nulls
kivake['DESCRIPTION_TRANSLATED'].fillna(' ', inplace=True)
kivake['TAGS'].fillna(' ', inplace=True)
kivake['LOAN_USE'].fillna(' ', inplace=True)
```

#### Create new columns of number of words and count of characters in the 3 text columns to use for modelling.  The text itself will be discarded from the Numeric models, but will be leveraged in the NLP models.

```python
#Step 2 is to create new columns of number of words and count of characters in the 3 text columns.
def word_len_count(column):
    word_count = len(column.split())
    return word_count

def char_len_count(column):
    char_count = column.replace(' ','')
    char_count = len(char_count[:])
    return char_count

kivake['word_count_DT'] = kivake['DESCRIPTION_TRANSLATED'].apply(word_len_count)
kivake['word_count_TAGS'] = kivake['TAGS'].apply(word_len_count)
kivake['word_count_LU'] = kivake['LOAN_USE'].apply(word_len_count)

kivake['char_count_DT'] = kivake['DESCRIPTION_TRANSLATED'].apply(char_len_count)
kivake['char_count_TAGS'] = kivake['TAGS'].apply(char_len_count)
kivake['char_count_LU'] = kivake['LOAN_USE'].apply(char_len_count)
```

```python
kivake.head(2)
```

```python
#Mark whether Image and Video was provided
kivake['IMAGE_ID'] = np.where(kivake['IMAGE_ID'].isnull(), 0, 1)
kivake['VIDEO_ID'] = np.where(kivake['VIDEO_ID'].isnull(), 0, 1)
```

```python
## Parse out borrower genders & borrower pictured by gender
kivake['FEM_COUNT'] = kivake['BORROWER_GENDERS'].str.count("female")
kivake['MALE_COUNT'] = kivake['BORROWER_GENDERS'].str.count("male")
kivake['PIC_TRUE_COUNT'] = kivake['BORROWER_PICTURED'].str.count("true")
kivake['PIC_FALSE_COUNT'] = kivake['BORROWER_PICTURED'].str.count("false")
```

```python
#Update with flag if there was any male or female in the dataset
kivake['ANY_FEM'] = kivake['BORROWER_GENDERS'].str.contains('female').map({True:1, False:0})
kivake['ANY_MALE'] = kivake['BORROWER_GENDERS'].str.contains('male').map({True:1, False:0})

#Update with month by extracting from the posted-time
kivake['month'] = kivake['POSTED_TIME'].str[5:7]
```

```python
#Drop unrequired columns

kivake.drop(columns = ['DESCRIPTION_TRANSLATED', 'LOAN_USE', 'TAGS', 'DESCRIPTION','LOAN_NAME', 'FUNDED_AMOUNT', 'DESCRIPTION', 'TOWN_NAME', 'CURRENCY_EXCHANGE_COVERAGE_RATE', 'CURRENCY', 'PARTNER_ID', 'PLANNED_EXPIRATION_TIME', 'DISBURSE_TIME', 'POSTED_TIME', 'RAISED_TIME', 'NUM_LENDERS_TOTAL', 'NUM_JOURNAL_ENTRIES', 'NUM_BULK_ENTRIES', 'BORROWER_NAMES', 'COUNTRY_NAME', 'CURRENCY_POLICY','year', 'BORROWER_GENDERS', 'BORROWER_PICTURED', 'IMAGE_ID', 'VIDEO_ID'], inplace=True)
```

```python
kivake.head(2)
```

```python
#Plots and correlation
#Plotting the bar plots for all categorical variables
def bar_plot(dataset, variable):
    var =dataset[variable]
    varValue = var.value_counts()
    plt.figure(figsize=(15,3))
    plt.bar(varValue.index, varValue,color=['#00008b','#00e5ee','#cd1076', '#008080','#cd5555','red','blue',])
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    
    plt.show()
    print("{}: \n {}".format(variable,varValue))
    
categoryc = ["STATUS", "ORIGINAL_LANGUAGE", "ACTIVITY_NAME", "SECTOR_NAME", "REPAYMENT_INTERVAL", "DISTRIBUTION_MODEL", "month", "ANY_FEM", "ANY_MALE", "LENDER_TERM"]

for c in categoryc:
    bar_plot(kivake, c)
```

```python
#Plotting histograms for numeric variables

def plot_hist(dataset, variable):
    plt.figure(figsize=(9,6))
    plt.hist(dataset[variable], bins=10,color='blue')
    plt.xlabel(variable)
    plt.ylabel("frequency")
    plt.title("{} distribution with hist".format(variable))
    plt.show()
    
numericVar = ["LOAN_AMOUNT","LENDER_TERM","word_count_DT", "char_count_DT", "word_count_LU", "char_count_LU", "word_count_TAGS", "char_count_TAGS", "FEM_COUNT", "MALE_COUNT"]
for n in numericVar:
    plot_hist(kivake, n)
```

```python
#Plotting correlation
fig, ax = plt.subplots(figsize=(13,13))         

# Sample figsize in inches
sns.heatmap(kivake.corr(), annot=True, linewidths=.5, ax=ax)
```

```python
#Reviewing groupby values for different categorical values
catcols = ['ORIGINAL_LANGUAGE', 'ACTIVITY_NAME', 'SECTOR_NAME', 'COUNTRY_CODE', 'REPAYMENT_INTERVAL', 'DISTRIBUTION_MODEL', 'month']

for i in range(1,len(catcols)):
    print(kivake.groupby(catcols[i], dropna=False)['STATUS'].agg(['count', 'mean']))
```

### There are a lot of activity names.  We need to consolidate before we make dummies

```python
#Reducing activity_name to some key attributes
cond = kivake['ACTIVITY_NAME'].value_counts()
threshold = 500
kivake['ACTIVITY_NAME'] = np.where(kivake['ACTIVITY_NAME'].isin(cond.index[cond >= threshold ]), kivake['ACTIVITY_NAME'], 'miscellaneous')

kivake['ACTIVITY_NAME'].value_counts()
```

```python
#Eliminating non-English language (Spanish & Russian)
kivake[kivake['ORIGINAL_LANGUAGE'] == 'English'] 
```

```python
#Creating interraction terms between character count and word count which showed up as highly correlated in the dataset
kivake['word_char_DT'] = kivake['word_count_DT'] * kivake['char_count_DT']
kivake['word_char_TAGS'] = kivake['word_count_TAGS'] * kivake['char_count_TAGS']
kivake['word_char_LU'] = kivake['word_count_LU'] * kivake['char_count_LU']

kivake['MALE_FEM'] = kivake['MALE_COUNT']*kivake['FEM_COUNT']
kivake['MALE_PIC'] = kivake['MALE_COUNT']*kivake['PIC_TRUE_COUNT']
kivake['FEM_PIC'] = kivake['FEM_COUNT']*kivake['PIC_TRUE_COUNT']
```

```python
#See how many nulls we have
kivake.isnull().sum()
```

```python
#Get mean/median for imputing
var_list = ['FEM_COUNT', 'MALE_COUNT', 'PIC_TRUE_COUNT', 'PIC_FALSE_COUNT', 'ANY_FEM', 'ANY_MALE', 'MALE_FEM', 'MALE_PIC', 'FEM_PIC']
 
[print(kivake[i].agg(['mean', 'median'])) for i in var_list]
```

```python
#Impute values here to be either MISSING or MEDIAN for numeric
fill_values = {'ORIGINAL_LANGUAGE' : 'MISSING', 'FEM_COUNT' : 1, 'MALE_COUNT' : 1,'PIC_TRUE_COUNT' : 1, 'PIC_FALSE_COUNT' : 0,'ANY_FEM' : 1,'ANY_MALE' : 1,'COUNTRY_CODE':'MISSING', 'MALE_FEM':1,'MALE_PIC':1,'FEM_PIC':1}
kivake.fillna(value = fill_values, inplace=True)
```

```python
#Creating file for model to work
kivake.to_csv('kivasmall.csv', index=False)
```
