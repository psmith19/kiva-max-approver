```python
#train-test split
X_train, X_test, y_train, y_test=train_test_split(X,
                                                 y,
                                                 test_size=.2,
                                                 stratify=y,
                                                 random_state=42)
```

### Run Through All Classifiers

```python
#create a list of the model classes to test
model_list = [
    LogisticRegression(),
    BaggingClassifier(),
    RandomForestClassifier(),
    ExtraTreesClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier()
]
```

```python
#vectorize features
cvec = CountVectorizer()
X_train_cvec = cvec.fit_transform(X_train)
X_test_cvec = cvec.transform(X_test)

#create a list to store each model's results in a dictionary
classifier_list = []

for model_obj in model_list:
    #instantiate each model 
    model = model_obj

    #fit the model
    model.fit(X_train_cvec, y_train) 
  
    #create a dictionary with scores and evaluation metrics for each model
    results_dict = {}    
    results_dict['model_name'] = str(model_obj)
    results_dict['train_score'] = model.score(X_train_cvec, y_train)
    results_dict['test_score'] = model.score(X_test_cvec, y_test)
    results_dict['cv_score'] = cross_val_score(model, X_train_cvec, y_train, cv = 5).mean()
        
    #add the dictionary to the list
    classifier_list.append(results_dict)

#create a dataframe and display dictionary
clf_results = pd.DataFrame(classifier_list)
clf_results
```

```python
kiva_full = pd.read_csv('/content/drive/MyDrive/Colab/group_project/pre_processed_kiva.csv').drop(columns = 'Unnamed: 0')
```

```python
kiva = kiva_full.groupby('STATUS').apply(lambda x: x.sample(frac=0.08))
```

```python
X = kiva['DESCRIPTION_TRANSLATED']
y = kiva['STATUS']

#train-test split
X_train, X_test, y_train, y_test=train_test_split(X,
                                                 y,
                                                 test_size=.2,
                                                 stratify=y,
                                                 random_state=42)
```

```python
y.value_counts()
```

```python
cvec = CountVectorizer()

# convert training data to dataframe
X_train_df_cvec = pd.DataFrame(cvec.fit_transform(X_train).todense(),
                         columns=cvec.get_feature_names())


# plot top occurring words
X_train_df_cvec.sum().sort_values(ascending = False).head(10).plot(kind='barh');
```

```python
#create a list of the model classes to test
model_list = [
    LogisticRegression(),
    BaggingClassifier(),
    RandomForestClassifier(),
    ExtraTreesClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    BernoulliNB()
]
```
